```python
#newTitle=[]
newTitle2=ratings['bname'].sort_values(ascending = True).unique()
#newTitle.append(newTitle2)
newTitle = pd.DataFrame(newTitle2)
newTitle.head()
```

```python
indices = pd.Series(newTitle.index, index = newTitle[0]).drop_duplicates()
indices.head()
```

```python
# cosine similarity를 바탕으로 비슷한 책 추천
def get_recommendations(bname, cosine_sim=cosine_sim):
    # 선택한 영화의 타이틀로부터 해당되는 인덱스를 받아와서
    # 선택한 영화를 가지고 연산
    idx = indices[bname]

    # 모든 영화에 대해서 해당 영화와의 유사도를 구함
    sim_scores = list(enumerate(cosine_sim[idx]))

    # 유사도에 따라 영화들을 정렬
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)

    # 가장 유사한 10개의 영화를 받아옴
    sim_scores = sim_scores[1:11]

    # 가장 유사한 10개의 영화의 인덱스를 받아옴
    movie_indices = [i[0] for i in sim_scores]

    # 가장 유사한 10개의 영화의 제목을 리턴
    return ratings['bname'].iloc[movie_indices]
```

```python
get_recommendations('13 Reasons Why')
```
