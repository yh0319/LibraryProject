```python
rname = ['userID', 'bname', 'rate']
ratings = pd.read_csv('ratings.csv', encoding = 'utf-8-sig', sep = '\t', header = None, names = rname)
ratings
```

```python
userRatings = ratings.pivot_table(index = ['userID'], columns = ['bname'], \
                              values = 'rate', fill_value = 0, dropna =False)
userRatings
```

```python
corrMatrix = userRatings.corr()
corrMatrix.head()
```

```python
corrMatrix = userRatings.corr(method='pearson', min_periods=100)
corrMatrix.head()
```

```python
simCandidates = pd.Series()
for i in range(0, len(userRatings.index)):
    print("adding sims for user" + str(userRatings.index[i]))
    sims = corrMatrix[userRatings.index[i]]
    sims = sims.map(lambda x: x * userRatings[i])
    simCandidates = simCandidates.append(sims)
    
print("sorting in decreasing order of similarity score: ")
simCandidates.sort_values(inplace = True, ascending = False)
print(simCandidates.head(10))
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
#사용자간 유사도
cosine_sim = pd.DataFrame(cosine_similarity(userRatings, userRatings))
cosine_sim.head()
```

```python
def get_recommendations(title, cosine_sim=cosine_sim):
    # 선택한 영화의 타이틀로부터 해당되는 인덱스를 받아옵니다. 이제 선택한 영화를 가지고 연산할 수 있습니다.
    idx = indices[title]

    # 모든 영화에 대해서 해당 영화와의 유사도를 구합니다.
    sim_scores = list(enumerate(cosine_sim[idx]))

    # 유사도에 따라 영화들을 정렬합니다.
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)

    # 가장 유사한 10개의 영화를 받아옵니다.
    sim_scores = sim_scores[1:11]

    # 가장 유사한 10개의 영화의 인덱스를 받아옵니다.
    movie_indices = [i[0] for i in sim_scores]

    # 가장 유사한 10개의 영화의 제목을 리턴합니다.
    return ratings['bname'].iloc[movie_indices]
```

```python
get_recommendations('후레자식')
```
