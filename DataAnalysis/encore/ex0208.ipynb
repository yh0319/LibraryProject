### user 테이블 만들기 
컬럼명 : userid, birth, sex<br>
데이터 수 : 442

```python
u1 = pd.read_csv('../title_rate/members_includingUnknown.csv', encoding = 'utf-8-sig')
u2 = u1.copy()
u2.head()
u21 = u2['id']
u23 = u2['sex']
u22 = u2['birth']
```

```python
df01 = pd.DataFrame(u21)
df02 = pd.DataFrame(u22)
df03 = pd.DataFrame(u23)
df04 = pd.concat([df01, df02, df03], axis = 1)
df04.head()
```

```python
df05 = df04.sort_values(["id"], ascending= True)
df05 = df05.reset_index(drop=True)
```

```python
df05.to_csv('user.csv', encoding = 'utf-8-sig', sep = '\t', index = False)
```
