```python
from collections import Counter
from konlpy.tag import Twitter
import pytagcloud
```

```python
f = open('word.txt', encoding='utf-8-sig')
```

```python
data = f.read()
```

```python
data
```

```python
nlp = Twitter()
nouns = nlp.nouns(data)
```

```python
nouns
```

```python
count =  Counter(nouns)
```

```python
count
```

```python
count.most_common(80)
```

```python
tag2 = count.most_common(80)
taglist = pytagcloud.make_tags(tag2, maxsize=80)
```

```python
del tag2[1]
```

```python
count.most_common(30)
```

```python
taglist = pytagcloud.make_tags(tag2, maxsize=80)
```

```python
taglist
```

```python
pytagcloud.create_tag_image(taglist, 'wordcloud.jpg', size=(900, 600), fontname='Korean', rectangular=False)
```
