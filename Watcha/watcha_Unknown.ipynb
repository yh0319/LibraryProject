```python
import time
import re
import sqlite3
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.by import By 
import csv
import pandas as pd
from selenium.common.exceptions import TimeoutException
from selenium.webdriver.support import expected_conditions as EC

# 익명의 사용자 계정 들어가기
def main():

    with open('watcha0125V2.csv', "r", encoding="utf-8") as f:
        addressList = []
        for i in f.readlines():
            i = i.replace('\ufeff', '')
            i = i.replace('\n', '')
            addressList.append(i)
            print('ccc')

        # 사용자별 도서 페이지 가기
        for j in range(len(addressList)):
            print(str(j)+'번째 사용자 도서페이지')
            driver = webdriver.Chrome("C:/driver/chromedriver.exe")
            address = addressList[j]
            url = 'https://watcha.com/en-KR/users/'+address+'/contents/books/'
            driver.get(url)

            time.sleep(2)
      
            # 사용자의 평점수 확인
            numRank = driver.find_element_by_xpath('//*[@id="root"]/div/div/section/section/section/div[1]/div/header/span').text
            numR = int(numRank)
            print(str(j)+'번째 사용자의 평점 수 : ' + str(numR))
            time.sleep(3)

            # 더보기 페이지 가기
            driver.find_element_by_xpath('//*[@id="root"]/div/div/section/section/section/div[1]/div/header/div/div/a').click()
            time.sleep(4)

            # 도서 목록 페이지에서 책 제목, 별점 가져오기위해 스크롤 내리기
            last_height = driver.execute_script("return document.body.scrollHeight")

            while True:

                # Scroll down to bottom
                driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")

                # Wait to load page
                time.sleep(2)

                # Calculate new scroll height and compare with last scroll height
                new_height = driver.execute_script("return document.body.scrollHeight")

                if new_height == last_height:
                    break

                last_height = new_height

            time.sleep(3)

            # 도서 목록 페이지에서 책 제목, 별점 가져오기
            soup = BeautifulSoup(driver.page_source, "lxml" )
            titless = soup.select('li > a')

            wTitle1 = []
            wRating1 = []
            for title in titless:
                reTitle = title.text
                Title2 = re.sub("★", "#~%", reTitle)
                Title3 = re.split("#~%", Title2)
                wTitle1.append(Title3[0])
                wRating1.append(Title3[-1])
                wTitle2 = wTitle1[1:]
                wRating2 = wRating1[1:]

            # 도서 상세페이지 열기
            wAuthor = []
            for b in range(numR):
                last_height = driver.execute_script("return document.body.scrollHeight")

                while True:
                    # 스크롤 내리기
                    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")

                    time.sleep(3)

                    # 새 스크롤 길이와 이전 스크롤 비교하기
                    new_height = driver.execute_script("return document.body.scrollHeight")

                    if new_height == last_height:
                        break
                    last_height = new_height

                    time.sleep(3)

                time.sleep(3)

                findurl = (driver.find_elements_by_xpath('//*[@id="root"]/div/div/section/section/div[1]/section/div[1]/div/ul/li[*]/a'))[b].get_attribute('href')
                driver.get(findurl)
                time.sleep(3)
                author = driver.find_element_by_xpath('//*[@id="root"]/div/div/section/div/div/div/section/div[2]/div/div/div/div/div[1]').text
                time.sleep(3)

                wAuthor.append(author)
                time.sleep(5)
                driver.back()

                print(str(b)+'번째 도서 정보 획득 완료')
                time.sleep(5)

            df01 = pd.DataFrame(wTitle2)
            df02 = pd.DataFrame(wAuthor)
            df03 = pd.DataFrame(wRating2)
            df04 = pd.concat([df01, df02, df03], axis = 1)              
            df04.to_csv('./watchaUnknown/unknown%d_%s.csv'%(j, i), sep ='\t', index = False, header=None, encoding = 'utf-8-sig')
            print(str(j)+'번째 익명의 사용자 csv 파일 생성 완료')

            time.sleep(3)
            driver.quit()
            time.sleep(3)
  
if __name__ == "__main__":
    main()
```
