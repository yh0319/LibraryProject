### - 크롤링한 각 데이터 에서 ISBN 번호 추출

## 연령별 인기 도서

```python
import pandas as pd
import numpy as np
import json
```

```python
with open('PopularBookByAge.json', encoding='utf-8-sig') as data_file: 
    byAge = json.load(data_file)
```

```python
def list_clear(list):
    id_list = []
    lists= []
    for x in (list):
        for y in x:
            if y['isbn'] and y['isbn'] not in id_list:
                id_list.append(y['isbn'])
                lists.append(y)
            else:
                pass
    return lists
```

```python
byAge = list_clear(byAge)
```

```python
byAge = pd.DataFrame.from_dict(byAge, orient='columns')
byAge
```

```python
ageISBN = byAge['isbn'].unique()
```

```python
ageISBN = pd.DataFrame(ageISBN)
ageISBN
```
