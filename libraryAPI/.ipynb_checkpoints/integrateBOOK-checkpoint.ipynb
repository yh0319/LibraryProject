## bookInfo1

```python
with open('json/bookInfo/bookInfo1.json', encoding='utf-8-sig') as data_file: 
    bookInfo1 = json.load(data_file)
```

```python
list = []
for i in range(len(bookInfo1)):
    for j in (bookInfo1):
        list.append(j[0])
```

```python
def clear_list(list):
    lists = []
    for i in range(len(list)):
        for j in (list):
            lists.append(j[0])
        
    return lists 
```

```python
bookInfo1= clear_list(bookInfo1)
```

```python
bookInfo1 = pd.DataFrame.from_dict(bookInfo1, orient='columns')
bookInfo1
```

```python
###  Author(저자)컬럼 간단하게 데이터 정제하기
```

```python
clear_author = []
for i in bookInfo['author']:
    print(i)
    i = i.replace('지은이:', '')
    i = i.replace('(지은이)', '')
    i = i.replace('글·그림:', '')
    i = i.replace('글·그림', '')
    i = i.replace(' 글, 그림: ', '') 
    i = i.replace('(글)', '')
    i = i.replace('글·각본:', '')
    i = i.replace('글:', '')
    i = i.replace('글', '')
    i = i.replace(' 글', '')
    i = i.replace('글,', '')
    i = i.replace('저자:', '')
    i = i.replace('그림: ', '')
    i = i.replace('지음', '')
    i = i.replace('·사진', '')
    i = i.replace('[지음]', '')
    i = i.replace('글,', '')
    i = i.replace(',', ';')
    i = i.replace(':', '')
    i = i.replace(':', ';')
    i = i.replace(' ', '')
    i = i.replace('[]', '')
    i = i.replace(' ', '')
    i = i.replace('...[등]', '')
    i = i.split(';')    
    print(i[0])
    print('----------------')
    bookInfo['author'].iloc[1]
    clear_author.append(i[0])


    
```

### 정규표현식으로 bookname 컬럼의 ':(콜론)이하 모든 문자열' 삭제하기  

```python
import re
```

```python
pattern = re.compile('^([^:]*)')
```

```python
test = bookInfo.copy()
```

```python
for i in range(1716):
    data = pattern.findall(test['bookname'].iloc[i])
    test['bookname'].iloc[i] = data[0].strip()

test
```
