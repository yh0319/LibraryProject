- 이 후 '}'   <- 블록잡고  ctrl+h, 한 후  '},' 아래 작성하고 replaceall 눌러서 전부 변경<br>
- 마지막으로 딕셔너리를 하나의 큰 리스트 [] 로 감싸주고 저장!!! 
- 그리고 아래 주피터 코드 실행해서 하면 됩니다! ㅠㅠㅠ

```python
import pandas as pd
import csv
import json
```

#### json 파일 합치기

```python
with open('sep1 - 복사본.json', encoding='utf-8-sig') as data_file: 
    sep1 = json.load(data_file)
```

```python
sep1 = pd.DataFrame(sep1)
```

```python
with open('sep2 - 복사본.json', encoding='utf-8-sig') as data_file: 
    sep2 = json.load(data_file)
```

```python
sep2 = pd.DataFrame(sep2)
```

```python
sep_merge = pd.concat([sep1,sep2])
```

```python
sep_merge
```

#### csv 파일 합치기

```python
w1 = pd.read_csv('sep1.csv', names=['bookname'])
```

```python
w2 = pd.read_csv('sep2.csv', names=['bookname'])
```

```python
w3 = pd.concat([w1, w2])
```

```python
w3
```

#### DataFrame 형태로 변환한 json, csv 파일을 합치기

```python
watchaYH = pd.concat([sep_merge, w3], axis=1)
```

```python
watchaYH
```

#### 타이틀 날리기

```python
del watchaYH['title']
```

```python
watchaYH
```

#### 최종 csv 파일로 저장

```python
watchaYH.to_csv('fianlYH.csv', sep='\t', encoding='utf-8', index=None)
```
