```python
with open('data.csv', 'r', encoding = 'utf-8') as f:
    datas = f.readlines()
```

```python
datas
```

```python
headlines = []
```

```python
import re
for i in range(len(datas)):
        if datas[i] != '\n':
            datas[i] = datas[i].replace('\ufeff', '')
            datas[i] = re.sub(r"'", "", datas[i])
            datas[i] = re.sub(r'"', '', datas[i])
            datas[i] = datas[i].replace('\n', '')
            datas[i] = datas[i].replace(',', '')
            datas[i] = datas[i].replace('.', ' ')
            datas[i] = datas[i].replace('…', ' ')
            datas[i] = datas[i].replace('“', '')
            datas[i] = datas[i].replace('”', '')
            datas[i] = datas[i].replace('‘', '')
            datas[i] = datas[i].replace('’', '')
            datas[i] = datas[i].replace('·', ' ')
            datas[i] = datas[i].replace('-', ' ')
            datas[i] = datas[i].replace('ㆍ', ' ')
            datas[i] = datas[i].replace('`', '')
            datas[i] = datas[i].replace('?', '')
            datas[i] = datas[i].replace('~', '')
            datas[i] = re.sub(r'\[(.*?)\]', '', datas[i])
            datas[i] = re.sub(r'\[(.*?)\]\s', '', datas[i])
            datas[i] = re.sub(r'\<(.*?)\>', '', datas[i])
            datas[i] = re.sub(r'\(\w*\)', '', datas[i])
            datas[i] = re.sub(r'^\s', '', datas[i])
            datas[i] = re.sub(r'\[.*', '', datas[i])
            datas[i] = re.sub(r'\(.*', '', datas[i])
            headlines.append(datas[i])
```

```python
import csv
with open('results.csv', 'w', encoding = 'utf-8') as f:
    csv_writer = csv.writer(f)
    for headline in headlines:
        csv_writer.writerow([headline])
```
